# Simple calculator

```python
from tkinter import *

root = Tk()
root.title('Idris Badmos: Demo Simple Calculator')
e = Entry(root, width= 50, borderwidth= 5, fg='black', bg='blue')
e.grid(row= 1, column=0 ,columnspan=3)



def click(number):
    current = e.get()
    e.delete(0, END)
    e.insert(0, str(current) + str(number))

def multiply():
    first_number = e.get()
    global f_num
    global math
    math = 'multiplication'
    f_num = int(first_number)
    e.delete(0, END)

def add():
    first_number = e.get()
    global f_num
    global math
    math = 'addition'
    f_num = int(first_number)
    e.delete(0, END)


def divide():
    first_number = e.get()
    global f_num
    f_num = int(first_number)
    global math
    math = 'division'
    e.delete(0, END)


def equal():
    second_number = e.get()
    e.delete(0, END)
    if math == 'multiplication':
        e.insert(0, f_num * int(second_number))
    if math == 'addition':
        e.insert(0, f_num + int(second_number))
    if math == 'subtraction':
        e.insert(0, f_num - int(second_number))
    if math == 'division':
        e.insert(0, f_num / int(second_number))



def clear():
    e.delete(0, END)

def subtract():
    first_number = e.get()
    global f_num
    f_num = int(first_number)
    global math
    math = 'subtraction'
    e.delete(0, END)



button_1 = Button(root, text='1', command=lambda: click(1), bg='red', padx=40, pady=10, borderwidth=5)
button_2 = Button(root, text='2', command=lambda: click(2), bg='red', padx=40, pady=10, borderwidth=5)
button_3 = Button(root, text='3', command=lambda: click(3), bg='red', padx=40, pady=10, borderwidth=5)

button_4 = Button(root, text='4', command=lambda: click(4), bg='red', padx=40, pady=10, borderwidth=5)
button_5 = Button(root, text='5', command=lambda: click(5), bg='red', padx=40, pady=10, borderwidth=5)
button_6 = Button(root, text='6', command=lambda: click(6), bg='red', padx=40, pady=10, borderwidth=5)

button_7 = Button(root, text='7', command=lambda: click(7), bg='red', padx=40, pady=10, borderwidth=5)
button_8 = Button(root, text='8', command=lambda: click(8), bg='red', padx=40, pady=10, borderwidth=5)
button_9 = Button(root, text='9', command=lambda: click(9), bg='red', padx=40, pady=10, borderwidth=5)

button_0 = Button(root, text='0', command=lambda: click(0), bg='red', padx=40, pady=10, borderwidth=5)
button_add = Button(root, text='+', command=add, bg='yellow', padx=15, pady=10, borderwidth=5)
button_multiply = Button(root, text='*', command=multiply, bg='yellow', padx=40, pady=10, borderwidth=5)
button_divide = Button(root, text='/', command=divide, bg='yellow', padx=15, pady=10, borderwidth=5)
button_equal = Button(root, text='=', command=equal, bg='grey', padx=40, pady=10, borderwidth=5)

button_subtract = Button(root, text='-', command=subtract, bg='yellow', padx=40, pady=10, borderwidth=5)
button_clear = Button(root, text='clear', command=clear, bg='yellow', padx=30, pady=10, borderwidth=5)

button_7.grid(row=2, column=0)
button_8.grid(row=2, column=1)
button_9.grid(row=2, column=2)

button_4.grid(row=3, column=0)
button_5.grid(row=3, column=1)
button_6.grid(row=3, column=2)

button_1.grid(row=4, column=0)
button_2.grid(row=4, column=1)
button_3.grid(row=4, column=2)

button_add.grid(row=5, column=1, sticky= E)
button_divide.grid(row=5, column=1, sticky=W)
button_multiply.grid(row=5, column=2)
button_0.grid(row=5, column=0)
button_equal.grid(row=6, column=2)

button_subtract.grid(row=6, column=0)
button_clear.grid(row=6, column=1)





root.mainloop()
```
