# Assignment: Explore how to Create Piecharts

```python
import matplotlib.pyplot as plt

# Data to plot
languages = ['python', 'java', 'C++', 'C#']
sizes = [1002.0, 1602.8, 605.7, 384.2]
colors = ['red', 'blue', 'cyan', 'magenta']
explode = (0,0.09,0,0)  # explode 1st slice

# Plot
fig1, ax1 = plt.subplots()
patches, texts, autotexts = ax1.pie(sizes, labels=languages, colors=colors, startangle=180, autopct='%1.1f%%')

for text in texts:
    text.set_color('red')
for autotext in autotexts:
    autotext.set_color('black')
    
ax1.axis('equal')
plt.tight_layout(pad=1.08, h_pad=100, w_pad=100, rect=(0,0,1,1))
plt.legend(languages,loc='best')
plt.show()



```

```python
import matplotlib.pyplot as plt
# Pie chart
languages = ['python', 'java', 'C++', 'C#']
sizes = [1002.0, 1602.8, 605.7, 384.2]
#colors
colors = ['red', 'blue', 'cyan', 'magenta']
 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, labels=languages, autopct='%1.1f%%', startangle=90)
#draw circle
centre_circle = plt.Circle((0,0),0.40,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```
