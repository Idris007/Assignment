# Simple interest Assignment

```python
# A code to find simple interest
'''
The parameters of simple interest are
1. Principal (P): This is the amount invested
2. Rate (R): interest rate
3. Simple interest (I): Unknown variable mostly
4. Time (T): Duration of investment

Formular
I = PRT/100
'''
# Define user input for P, R, & T
Principal= int(input('Enter the principal here: '))
Rate= int(input('Enter the rate here: '))
Time= int(input('Enter time here: '))
Currency= input('Enter currency here: ')
# Function to calculate simple interest with parameters P, R, & T
def simple_interest(Principal, Rate, Time):
    Interest= (Principal * Rate * Time)/100
    return int(Interest)
print(f'{simple_interest(Principal, Rate, Time)} {Currency} ')
```
